```python
import moviepy.editor,moviepy,PIL
import cv2
import os
```

```python
videoname="ed.mkv"
foldername=videoname.split(".")[0]
print(foldername)
```

```python
if not os.path.isdir(foldername):
    os.mkdir(foldername)
    
clip=moviepy.editor.VideoFileClip(videoname)
clip.write_images_sequence(f"{foldername}/%d.jpg",fps=clip.fps,verbose=False,logger=None)
```
